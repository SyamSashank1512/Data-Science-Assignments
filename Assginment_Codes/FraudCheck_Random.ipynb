```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
num_trees = 200
#max_features = 3
kfold = KFold(n_splits=10, random_state=7, shuffle=True) #Bootstrap

model = RandomForestClassifier(n_estimators=num_trees,max_depth=5,  min_samples_leaf=5,max_features='sqrt')
results = cross_val_score(model, x, y, cv=kfold)
print(results.mean())
```

```python
#AdaBoostClassifier
from sklearn.ensemble import AdaBoostClassifier

kfold = KFold(n_splits=10, random_state=7,shuffle=True) 
model1 = AdaBoostClassifier(n_estimators=100, random_state=7)
results = cross_val_score(model1, x, y, cv=kfold)
print(results.mean())
```

```python
from numpy import loadtxt
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
Y_train = le.fit_transform(Y_train)
```

```python
model = XGBClassifier(n_estimators=90,max_depth=3)
model.fit(X_train, Y_train)
```

```python
y_pred_test = model.predict(X_test)
predictions = [round(value) for value in y_pred_test]
predictions
```

```python
accuracy = accuracy_score(Y_test, predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```

```python
import lightgbm as lgb
d_train = lgb.Dataset(X_train, label=Y_train)
```

```python
params = {}
params['learning_rate'] = 0.002, # 0.1,0.3
params['boosting_type'] = 'gbdt' # dart
params['objective'] = 'binary'
params['metric'] = 'binary_logloss'
params['sub_feature'] = 0.5 
params['num_leaves'] = 10
params['min_data'] = 50
params['max_depth'] = 10
```

```python
params
```

```python
clf = lgb.train(params, d_train, 1400)
```

```python
y_pred = clf.predict(X_test)
```

```python
predictions = [round(value) for value in y_pred]
```

```python
accuracy = accuracy_score(Y_test, predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```
