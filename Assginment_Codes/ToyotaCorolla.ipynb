```python
# MultiCollinearity

from statsmodels.stats.outliers_influence import variance_inflation_factor as VIF
```

```python
def Cal_Vif(x):
    
    #calculating vif values
    vif=pd.DataFrame()
    vif["Variables"]=x.columns=["Age_08_04","KM","HP","cc","Doors","Gears","Quarterly_Tax","Weight"]
    vif["Cal_Vif"]=[VIF(x.values , i) for i in range(x.shape[1])]
    
    return(vif)
    
```

```python
x1=x.iloc[:,:-1]
Cal_Vif(x)
```

#  Model Delection Diagnostics

```python
#Cooks Distance

from statsmodels.graphics.regressionplots import influence_plot
```

```python


influence=model1.get_influence()  # Create the variable for the model1

(c, _) = influence.cooks_distance  #Creating cooks variable 
print(c, _)

#visualizing the cook's Distance

fig = plt.subplots(figsize=(20, 7))
plt.stem(np.arange(len(df)), np.round(c, 3))
plt.xlabel('Row index')
plt.ylabel('Cooks Distance')
plt.show()
```

```python
# Hat-Points/ Leverage /Influence

k=df.shape[1]
n=df.shape[0]

leverage_cutoff=3*((k+1)/n)
leverage_cutoff
```

```python
# Plotting Influence 

influence_plot(model1, alpha=0.5)

y1=[i for i in range(-2,8)]
x1=[leverage_cutoff for i in range(10)]

plt.plot(x1,y1,'r+')
plt.show()
```
