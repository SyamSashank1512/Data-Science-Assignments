```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
freq_pattern=apriori(df,min_support=0.05,use_colnames=True)
freq_pattern.head(50)
```

```python
freq_i=association_rules(freq_pattern, metric='lift', min_threshold=1)
freq_i
```

```python
freq_i[["support","confidence","lift"]].hist(figsize=(15,7))
plt.show()
```

```python
freq_i.sort_values(by='lift')
```

```python
#Low Support 
low_support=0.2

freq_pattern1=apriori(df, min_support=low_support, use_colnames=True)
freq_pattern1.head(50)
```

```python
#Low Confidence

low_confidence=0.2

freq_items1=association_rules(freq_pattern1, min_threshold=low_confidence, metric="lift")
freq_items1
```

```python
#visualization of Support and Confidence

freq_items1[["support","confidence","lift"]].hist(figsize=(15,7))
plt.show()
```

```python
freq_items1.sort_values(by="lift")   #Here,the low support and low confidence will have the not meaningful data due to low threshold 
```
