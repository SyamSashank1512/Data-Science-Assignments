```python
from sklearn.preprocessing import LabelEncoder
```

```python
labelencoder= LabelEncoder()

df["ShelveLoc"]=labelencoder.fit_transform(df["ShelveLoc"])

df["Urban"]=labelencoder.fit_transform(df["Urban"])

df["US"]=labelencoder.fit_transform(df["US"])
```

```python
df
```

```python
X=df.drop(columns="Sales")
X.head()
```

```python
Y=df[["Sales"]]
Y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=20)
```

```python
X_train
```

```python
y_train=labelencoder.fit_transform(Y_train)
y_train
y_test= labelencoder.fit_transform(Y_test)
print(y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import tree
```

```python
model = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred_train=model.predict(X_train)
y_pred_train
```

```python
y_pred_test=model.predict(X_test)
y_pred_test
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_pred_test,y_test))
```

```python
from sklearn.metrics import accuracy_score

print("Accuracy:", accuracy_score(y_train, y_pred_train))
```

```python
from sklearn.metrics import confusion_matrix

print("\nConfusion Matrix:\n", confusion_matrix(y_train, y_pred_train))
```

```python
y=labelencoder.fit_transform(Y)
y
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
num_trees = 200
#max_features = 3
kfold = KFold(n_splits=10, random_state=7, shuffle=True) #Bootstrap

model = RandomForestClassifier(n_estimators=num_trees,max_depth=5,  min_samples_leaf=5,max_features='sqrt')
results = cross_val_score(model, X, y, cv=kfold)
print(results.mean())
```

```python
#AdaBoostClassifier
from sklearn.ensemble import AdaBoostClassifier

kfold = KFold(n_splits=10, random_state=7,shuffle=True) 
model1 = AdaBoostClassifier(n_estimators=100, random_state=7)
results = cross_val_score(model1, X, y, cv=kfold)
print(results.mean())
```

```python
import lightgbm as lgb
d_train = lgb.Dataset(X_train, label=Y_train)
```
