```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder=LabelEncoder()
```

```python
train_df["workclass"]=label_encoder.fit_transform(train_df["workclass"])

train_df["education"]=label_encoder.fit_transform(train_df["education"]) 

train_df["maritalstatus"]=label_encoder.fit_transform(train_df["maritalstatus"]) 

train_df["occupation"]=label_encoder.fit_transform(train_df["occupation"]) 

train_df["relationship"]=label_encoder.fit_transform(train_df["relationship"]) 

train_df["race"]=label_encoder.fit_transform(train_df["race"]) 

train_df["sex"]=label_encoder.fit_transform(train_df["sex"]) 

train_df["native"]=label_encoder.fit_transform(train_df["native"])

train_df["Salary"]=label_encoder.fit_transform(train_df["Salary"])
```

```python
test_df=pd.read_csv(r'C:\Users\SHYAM SASHANK\OneDrive\Desktop\Data_Science\Assginment_CSV\SalaryData_Test_navies.csv')
```

```python
test_df["workclass"]=label_encoder.fit_transform(test_df["workclass"])

test_df["education"]=label_encoder.fit_transform(test_df["education"]) 

test_df["maritalstatus"]=label_encoder.fit_transform(test_df["maritalstatus"]) 

test_df["occupation"]=label_encoder.fit_transform(test_df["occupation"]) 

test_df["relationship"]=label_encoder.fit_transform(test_df["relationship"]) 

test_df["race"]=label_encoder.fit_transform(test_df["race"]) 

test_df["sex"]=label_encoder.fit_transform(test_df["sex"]) 

test_df["native"]=label_encoder.fit_transform(test_df["native"])

test_df["Salary"]=label_encoder.fit_transform(test_df["Salary"])
```

```python
test_df
```

```python
# Here, the train data is taken 

array = train_df.values
X_train= array[:,0:13]
Y_train = array[:,13]

X_train
Y_train
```

```python
# Here, the test data is taken

array = test_df.values
X_test= array[:,0:13]
Y_test= array[:,13]

X_test
Y_test
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
nb_classifier=GaussianNB()
```

```python
nb_classifier.fit(X_train,Y_train)
```

```python
y_pred_train = nb_classifier.predict(X_train)
y_pred_train
```

```python
y_pred_test = nb_classifier.predict(X_test)
y_pred_test
```

```python
nb_classifier.predict([[33,2,9,13,4,9,3,4,1,0,0,40,37]])
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
accuracy=accuracy_score(Y_test,y_pred_test)
print(accuracy)
```

```python
print(classification_report(Y_test,y_pred_test))
```

```python
print(confusion_matrix(Y_test,y_pred_test))
```

```python
import seaborn as sns

cm = confusion_matrix(Y_test, y_pred_test)

# Plot the confusion matrix as a heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
