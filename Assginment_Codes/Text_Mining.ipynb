```python
df=elon.drop(columns=["Unnamed: 0"])
df
```

```python
df1 = [t.strip() for t in df.Text]  # Removing the extra spcaes in the df.Text '  Hello  ' ----> 'Hello'
df1 = [Text for Text in df1 if Text]
df1
```

```python
 # Now, Join the df1
text = ' '.join(df1)
text
```

```python
# For removing the Punctuations

no_punc_text = text.translate(str.maketrans('', '', string.punctuation))
```

```python
no_punc_text # It also removes the special symbols like "@", "!", "links also to words https://t.co/Fc9uhQSd7O--->httpstco9WFKMYu2oj"
```

```python
import nltk
nltk.download('punkt')
```

```python
# Now, the no_punc_text is converted into tokens

import nltk
from nltk.tokenize import word_tokenize
token_text= word_tokenize(no_punc_text)
token_text
```

```python
# Now, the token_text are used and removing the stop_word

from nltk.corpus import stopwords
# Here, there is a stopword txt file 
#importing the file
with open(r"C:\Users\SHYAM SASHANK\OneDrive\Desktop\Data_Science\Assginment_CSV\stop.txt") as file:
    contents = file.read()
```

```python
words = contents.split()
```

```python
words
    
```

```python
# Here, removing of the stop words is done where the word is a variable in which it helps in all token_text compares with the words 
        
no_stop_tokens = [word for word in token_text if not word in words]
```

```python
no_stop_tokens
```

```python
#Normalize the data
lower_words = [x.lower() for x in no_stop_tokens]
```

```python
lower_words
```

```python
# Importing the spacy for lemmatizer purpose
import spacy
# Loading English language small model of spacy 
nlp = spacy.load('en_core_web_sm') 
```

```python
# Now, Loading a variable "doc" we use join keyword for the nlp 
doc = nlp(' '.join(no_stop_tokens))
```

```python
# Print of Doc
doc
```

```python
lemmas = [token.lemma_ for token in doc]
lemmas
```
