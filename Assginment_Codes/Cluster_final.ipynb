```python
# Dropping columns who are having null values more than 30%

# Calculating the percentage of null values for each column
null_percentage = (df.isnull().mean() * 100).round(2)

# Identify columns with more than 30% null values
columns_to_drop = null_percentage[null_percentage > 30].index
print('Columns having null values more than 30% are :-\n',columns_to_drop,'\n\n')

# Drop columns with more than 30% null values
df1= df.drop(columns=columns_to_drop)
```

```python
# 5.Checking for NAN Values in Dataset in Column Wise

df1.isna().any()
```

```python
# Replacing of NAN Values In dataset.

df2=df1.replace(np.nan,0)
df2
```

```python
# Checking for NAN Values

df2.isna().any()
```

```python
# Dropping of Columns in DataSet

x= df2.drop(columns=["Population 15-64","Number of Records","Population 65+","Population 0-14"])
x
```

```python
# Converting the "GDP","Tourism Inbound" ,"Health Exp/Capita" and "Tourism Outbound" as Floats Values.

x["GDP"] = x["GDP"].replace('[\$,]','',regex=True).astype(float)
x["Tourism Inbound"] = x["Tourism Inbound"].replace('[\$,]','',regex=True).astype(float)
x["Tourism Outbound"] = x["Tourism Outbound"].replace('[\$,]','',regex=True).astype(float)
x["Health Exp/Capita"] = x["Health Exp/Capita"].replace('[\$,]','',regex=True).astype(float)

x
```

```python
# Converting "Country" Categorical to Numerical data

from sklearn.preprocessing import LabelEncoder

labelencoder_ = LabelEncoder()
x["Country"] = labelencoder_.fit_transform(x["Country"])
x
```

```python
# Feature Scaling

from sklearn.preprocessing import MinMaxScaler
norm = MinMaxScaler()
df_norm = norm.fit_transform(x)

df_norm
```

Here, Feature Scaling is done as Machine Learning is Sensitive. So, that Feature Scaling plays a major role in Clustering.

## Buliding Model For Dataset 

### 1. KMeans Clustering 

```python
# For finding the NO.OF CLUSTERS we need to Plot Elbow Graph

from sklearn.cluster import KMeans
wcss=[]

for i in range(1,10):
    kmeans=KMeans(n_clusters=i,random_state=20)
    kmeans.fit(df_norm)
    wcss.append(kmeans.inertia_)

plt.title("Elbow Graph")
plt.plot(range(1,10),wcss)
plt.xlabel("No.of Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
#Building the Cluster

k_cluster=KMeans(5,random_state=20)
k_cluster.fit(df_norm)
```

```python
# Printing k_cluster.labels_

k_cluster.labels_
```

```python
Y=pd.DataFrame(k_cluster.labels_)
```

```python
x[["KMeans"]]=Y
x
```

```python
# For the finding Quality of the Clusters

from sklearn.metrics import silhouette_score

score =silhouette_score(x, kmeans.labels_ , metric='euclidean')
score
```

### 3. DBSCAN 

```python
import seaborn as sbn
from sklearn.cluster import DBSCAN
```

```python
from sklearn.neighbors import NearestNeighbors

nearest_neighbors = NearestNeighbors(n_neighbors=11)
neighbors = nearest_neighbors.fit(df_norm)             #Fitting the df_norm in nearest neighbors

distances, indices = neighbors.kneighbors(df_norm)     #Finding of the distance and indices of the df_norm

print(distances[0,10])
print(indices)

distances = np.sort(distances[:,10], axis=0)   #Here the placing data points for 11th data point distance
print("\n The distance of the data points which is nearest to 11th point\n",distances)
fig = plt.figure(figsize=(5, 5))
plt.plot(distances)
plt.xlabel("Points")
plt.ylabel("Distance")   
```

```python
#Plotting of the eps we use the knee plot using the kneed library

from kneed import KneeLocator     #importing the knee plot

i=np.arange(len(distances))    # 'i' is a variable where it contains the distance

# Implementing the knee method on distance
knee= KneeLocator(i, distances, S=1, curve='convex', direction='increasing', interp_method='polynomial')
# Here, the 'i' is the array represent the X-axis
# Distance is the Y-axis and the dataset
# Sentivity(S) is a parameter used to control senstivity of the knee detection of the data, S larger than less senstivity 
# Curve is type of curve should be fitting to the data. Here, the convex due there is a curve near the highest point in graph.
# Direction is a parameter in which the serach should be done either increasing (or) decreasing. Here, the increasing is used due to it is increasing plot.
# Interp_method is a interpolation method for dataset to plot the knee point, here it is polynomial

fig = plt.figure(figsize=(5, 5))
knee.plot_knee()
plt.xlabel("Points")
plt.ylabel("Distances")

print(distances[knee.knee])
```

```python
# Now considering the Eps and the Minpts

dbscan= DBSCAN(eps=0.3, min_samples=15)
cluster=dbscan.fit_predict(df_norm)
```
