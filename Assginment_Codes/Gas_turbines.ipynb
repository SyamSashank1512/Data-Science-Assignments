```python
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
import numpy
```

```python
X=df.drop(columns="TEY")
X.head()
```

```python
Y=df[["TEY"]]
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test= train_test_split(X, Y, random_state=30, test_size=0.4)
```

```python
X_train
```

```python
X_test
```

```python
Y_train
```

```python
Y_test
```

```python
# create model
model = Sequential()
model.add(Dense(20, input_dim=10,  activation='relu')) #1st layer
model.add(Dense(10,  activation='relu')) #2nd layer
model.add(Dense(10,  activation='relu')) #3nd layer 
model.add(Dense(1, activation='sigmoid')) #4th layer or output layer
```

```python
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
gas=model.fit(X_train, Y_train, validation_data=(X_test,Y_test), epochs=100, batch_size=20)
```

```python
scores = model.evaluate(X, Y)
print(model.metrics_names)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
print(scores)
```

```python
loss, accuracy = model.evaluate(X_test, Y_test)
print(f'Test Loss: {loss:.4f}, Test Accuracy: {accuracy:.4f}')
```
