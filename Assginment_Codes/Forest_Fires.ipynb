```python
from sklearn.preprocessing import LabelEncoder 
```

```python
label_encoder=LabelEncoder()
```

```python
df["month"]=label_encoder.fit_transform(df["month"])

df["day"] = label_encoder.fit_transform(df["day"])

df["size_category"] = label_encoder.fit_transform(df["size_category"])
```

```python
df
```

```python
import seaborn as sns
sns.heatmap(df.corr())
```

```python
X =df.drop(columns= ("size_category"))
X
```

```python
y=df[["size_category"]]
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, y, random_state=30, test_size=0.4)
```

```python
X_train
```

```python
X_test
```

```python
Y_train
```

```python
Y_test
```

```python
from sklearn.svm import SVC
svm = SVC(kernel='rbf')
svm.fit(X_train,Y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred_train= svm.predict(X_test)
a=accuracy_score(Y_test, y_pred_train) * 100
print("Accuracy:", a)
```

```python
y_pred_train
```

```python
from sklearn.metrics import classification_report

print(classification_report(Y_test,y_pred_train))
```

```python
# Now, finding the "Linear"

linear = SVC(kernel='linear',C=0.01)
linear.fit(X_train,Y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred_train1= linear.predict(X_test)
a1=accuracy_score(Y_test, y_pred_train) * 100
print("Accuracy:", a1)
```

```python
y_pred_train1
```

```python
print(classification_report(Y_test,y_pred_train1))
```
