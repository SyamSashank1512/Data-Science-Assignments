```python
books_df= df1.pivot_table(index='UserID' , columns='BookTitle' , values='BookRating')

books_df
```

```python
books_df.fillna(0, inplace=True)
```

```python
books_df
```

```python
# Calculating the Cosine similarity 

from sklearn.metrics import pairwise_distances as pd
from scipy.spatial.distance import cosine, correlation
```

```python
books_cos = 1-pd(books_df.values,metric='cosine')  

# Here, the cosine will calculate between the books_names i and j in the dataset  

# Here, the pairwise will compute the pairwise distances between the books_name books_df
# Here, the 1- pd is used because whenever there is a dissimilarity(-1) it will represented as 0 instead of -1,0 means no similarity

books_cos
```

```python
# Converting the books_cos as Data Frame
import pandas as pd

df_cos = pd.DataFrame(books_cos)

df_cos
```

```python
# Converting the UserID as Rows and Columns for the df_cos

df_cos.index = books_df1.userid.unique()     #For Rows
df_cos.column = books_df1.userid.unique()    #For Columns
```

```python
df_cos
```

```python
df_cos_array = df_cos.values #Converting the df_cos to array for using the fill_diagonal

# Fill the diagonal with zeros using NumPy
np.fill_diagonal(df_cos_array, 0)

# Convert the modified NumPy array back to a DataFrame
df_cos = pd.DataFrame(df_cos_array, columns=df_cos.columns, index=df_cos.index)
```
