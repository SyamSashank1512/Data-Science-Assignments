```python
# By using the label encoding we have to convert the train_df string into Numeric data

from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder=LabelEncoder()
```

```python
# Now we are trying to change the workclass(String) into Integer data

train_df["workclass"]=label_encoder.fit_transform(train_df["workclass"])
```

```python
train_df
```

```python
# Now the same applied to the all the features in the train_df

train_df["education"]=label_encoder.fit_transform(train_df["education"]) 

train_df["maritalstatus"]=label_encoder.fit_transform(train_df["maritalstatus"]) 

train_df["occupation"]=label_encoder.fit_transform(train_df["occupation"]) 

train_df["relationship"]=label_encoder.fit_transform(train_df["relationship"]) 

train_df["race"]=label_encoder.fit_transform(train_df["race"]) 

train_df["sex"]=label_encoder.fit_transform(train_df["sex"]) 

train_df["native"]=label_encoder.fit_transform(train_df["native"])

train_df["Salary"]=label_encoder.fit_transform(train_df["Salary"])
```

```python
# Using the train_df as X_train and Y_train

array = train_df.values
X_train= array[:,0:13]
Y_train = array[:,13]


Y_train
```

```python
# Now, the test_df as X_test and Y_test

array = test_df.values
X_test = array[:,0:13]
Y_test= array[:, 13]

Y_test
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV , RandomizedSearchCV
```

```python
# Here, finding the different types kernels like 'rbf', "Linear","Polynomial", "Sigmoid"

svm = SVC(kernel='rbf')
svm.fit(X_train,Y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred_train= svm.predict(X_test)
a=accuracy_score(Y_test, y_pred_train) * 100
print("Accuracy:", a)
```

```python
y_pred_train
```

```python
from sklearn.metrics import classification_report

print(classification_report(Y_test,y_pred_train))
```
